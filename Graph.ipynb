```
import matplotlib.pyplot as plt
import numpy as np
data= {'a':np.arange(30),'c':np.random.randint(0,30,30),'d':np.random.randn(30)}
data['b']=data['a']+ 15 +np.random.randn(30)
data['d']=np.abs(data['d'])*200
plt.scatter('a','b',c='c',s='d',data=data)
plt.show()
```
